# What are top  10 highest-rated iphones on Flipkart in India

```python
highest_rated = data.sort_values(by = ['Star Rating'],ascending = False)

highest_rated = highest_rated.head (10)

print (highest_rated['Product Name'])
```

# How Many rating do highest reated iphones on Flipkart Have

```python
# Calculate the counts for 'Product Name'
iphones = highest_rated['Product Name'].value_counts()

# Extract labels and counts
labels = iphones.index
counts = highest_rated ['Number Of Ratings']  # Correct way to get counts from value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='skyblue')  # Create bar chart
plt.xlabel('Product Name')
plt.ylabel('Number of Ratings')
plt.title('Number of Ratings for Highest Rated iPhones')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()  # Display the chart

```

# Which Iphone Has the highest number of reviews on Flipkart

```python
iphones = highest_rated['Product Name'].value_counts()

# Extract labels and counts
labels = iphones.index
counts = highest_rated ['Number Of Reviews']  # Correct way to get counts from value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='black')  # Create bar chart
plt.xlabel('Product Name')
plt.ylabel('Number Of Reviews')
plt.title('number Of reviews for Highest Rated iPhones')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()  # Display the chart
```

# What is relationship between sale price of iphone and number of rating on Flipkart

```python

# Ensure there are no extra spaces in column names
data.columns = data.columns.str.strip()

# Create a scatter plot with a trendline
plt.figure(figsize=(8, 6))

# Use sns.regplot to add a regression line
sns.regplot(
    data=data,  # Pass the DataFrame here
    x="Number Of Ratings",  # Ensure this column name exists exactly
    y="Sale Price",  # Column for Sale Price
    scatter_kws={"s": 100, "alpha": 0.7},  # Customizing the scatter points
    line_kws={"color": "red", "lw": 2},  # Customizing the trendline color and width
    ci=None,  # Remove confidence interval around the trendline
    robust=True,  # Makes the line less sensitive to outliers
)

# Add title and axis labels
plt.title('Relationship between Sales Price and Number of Ratings of iPhones', fontsize=14)
plt.xlabel('Number of Ratings', fontsize=12)
plt.ylabel('Sale Price (in INR)', fontsize=12)

# Show the plot
plt.show()

```

There is negative line relationship between sale price of Iphone and number of rating .
it mean iphone with lower sales price are sold more in India

# What is relationship between discount percentage and number of ratings of iphone on Flipkart

```python
# Ensure there are no extra spaces in column names
data.columns = data.columns.str.strip()

# Create a scatter plot with a trendline
plt.figure(figsize=(8, 6))

# Use sns.regplot to add a regression line
sns.regplot(
    data=data,  # Pass the DataFrame here
    x="Number Of Ratings",  # Ensure this column name exists exactly
    y="Discount Percentage",  # Column for Sale Price
    scatter_kws={"s": 100, "alpha": 0.7},  # Customizing the scatter points
    line_kws={"color": "red", "lw": 2},  # Customizing the trendline color and width
    ci=None,  # Remove confidence interval around the trendline
    robust=True,  # Makes the line less sensitive to outliers
)

# Add title and axis labels
plt.title('Relationship between Sales Price and Number of Ratings of iPhones', fontsize=14)
plt.xlabel('Number of Ratings', fontsize=12)
plt.ylabel('Discount Percentage', fontsize=12)

# Show the plot
plt.show()
```

# figure the least expensive and most expensive iphones in India Market, along all specification

```python
# Find the row corresponding to the most expensive product
most_expensive = data.loc[data['Sale Price'].idxmax()]

# Find the row corresponding to the least expensive product
least_expensive = data.loc[data['Sale Price'].idxmin()]

# Display the result
print("Most Expensive Product:")
print(most_expensive)

print("\nLeast Expensive Product:")
print(least_expensive)
```
